## Moving Average

```python
Moving_Average = 삼성전자['날짜'].to_frame()

for stock in stocks:
    Moving_Average= pd.concat([Moving_Average, stock.loc[:, ['MA']] ], axis=1)
```

```python
Moving_Average = Moving_Average.set_index('날짜')
Moving_Average.columns= ['삼성전자','SK하이닉스','NAVER','씨젠','우리들휴브레인','현대차','DGB금융지주','미스터블루','셀트리온','데일리블록체인','소리바다','한화솔루션','아모레퍼시픽','CJ대한통운','GS건설']
Moving_Average
```
